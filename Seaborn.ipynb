```python
import seaborn as sns 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
fmri=sns.load_dataset("fmri")
fmri.to_csv("fmri-data")
sns.relplot(data=fmri,kind="line",x="timepoint",y="signal",col="region",hue="event",style="event")
```

```python
tips=sns.load_dataset('tips')
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Total Bill")
plt.show()
```

```python
tips=sns.load_dataset('tips')
sns.lineplot(x='sex',y='tip',data=tips,hue='day')
plt.title("Gender and Tip Relationship")
plt.show()
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
plt.title(" Tota Bill")
plt.show()
```

```python
sns.lineplot(x='day',y='total_bill',data=tips,hue='sex')
plt.title('Line plot: Total bill by day(Differentiated by gender)')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Box plot: Total bill by day')
plt.show()
```

```python
sns.violinplot(x='day',data=tips,y='total_bill',hue='sex')
plt.title("Total Bill Per Day")
plt.show()
sns.violinplot(x='day',data=tips,y='total_bill')
plt.title("Total Bill Per Day")
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.suptitle("Pair Plot",y=1.02)
plt.show()
```

```python
iris=sns.load_dataset('iris')
correlation_matrix=iris.select_dtypes(include='number').corr()
sns.heatmap(correlation_matrix,annot=True)
```
