```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,30,100)
y=np.sin(x)
plt.plot(x,y,color='red',label='sin wave',linewidth=4)
# Label
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Plot Title
plt.title("My First Plot")

# Plot Legend
plt.legend()
plt.show()
```

```python
x=np.linspace(1,30,300)
y=np.cos(x)
plt.plot(x,y,color='red',label='sin wave',linewidth=4)
# Label
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Plot Title
plt.title("My First Plot")

# Plot Legend
plt.legend()
plt.show()
```
